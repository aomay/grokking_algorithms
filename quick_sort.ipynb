```python
import random
```

```python
def  quick_sort(arr):
    if len(arr) < 2:
        return arr
    pivot = arr.pop(random.randrange(len(arr)))
    less = [i for i in arr if i <= pivot]
    gr = [i for i in arr if i > pivot]
    return quick_sort(less) + [pivot] + quick_sort(gr)
```

```python
lst = [random.randrange(10) for i in range(10)]
print(lst)
quick_sort(lst)
```
