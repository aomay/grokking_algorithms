```python
def sum_els(arr):
    if arr == []:
        return 0
    else:
        return arr[0] + sum_els(arr[1:])
```

```python
def count_els(arr):
    if arr == []:
        return 0
    return 1 + count_els(arr[1:])
```

```python
def max_el(arr):
    if count_els(arr) == 1:
        return arr[0]
    if arr[0] > max_el(arr[1:]):
        return arr[0]
    return max_el(arr[1:])
```

```python
def bin_search(arr, num):
    mid = int(count_els(arr)/2)
    if arr[mid] == num:
        return mid
    if arr[mid] > num:
        return bin_search(arr[:mid], num)
    return mid + bin_search(arr[mid:], num)
```
