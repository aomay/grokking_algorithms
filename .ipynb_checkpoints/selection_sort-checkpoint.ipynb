```python
def find_smallest(arr):
    smallest = arr[0]
    smallest_id = 0
    for id in range(len(arr)):
        if arr[id] < smallest:
            smallest = arr[id]
            smallest_id = id
    return smallest_id

def selection_sort(arr):
    sorted = []
    while len(arr) != 0:
        smallest_id = find_smallest(arr)
        sorted.append(arr.pop(smallest_id))
    return sorted
```

```python
selection_sort([2, 7, -3, 0, 1])
```
